```python
# 단, 소멸 대상이 되었다고 바로 소멸되는 것은 아니다.
# Python은 Garbage Collection을 통해 메모리를 관리한다.
# object를 reference하는 변수의 수를 가리켜 'reference count'라고 한다.
# reference count가 0이 되면 해당 object는 소멸 대상이 된다. 

r1 = [1, 2, 3]   # list [1, 2, 3]의 reference count = 1
r2 = r1          # list [1, 2, 3]의 reference count = 2

r1 = 'simple'    # list [1, 2, 3]의 reference count = 1
r2 = 'happy'     # list [1, 2, 3]의 reference count = 0 -> Garbage Collection 대상
```

* Python은 Garbage Collection을 통해 메모리를 관리한다고 했는데, 저 자세히 하면 PVM(Python Virtual Machine)이 관리한다.
   *  python interpreter는 PVM의 구현 방식에 따라 종류가 다양하다.
   *  PVM이 C언어로 구현되어 있는 가장 일반적이고 널리 사용되는 interpreter를 CPython이라고 한다.
   *  PVM이 JAVA로 구현되어 있는 interpreter를 Jython이라고 한다.
   *  PVM이 MicroSoft사의 .NET으로 구현되어 있는 interpreter를 IronPython이라고 한다.

```python
import sys

class Simple:
    pass

class Simple2:
    pass

s1 = Simple()  # refcount is started from 1 (name variable)
s2 = Simple2() # refcount is started from 1 (name variable)
print(sys.getrefcount(s1)) # 1 + 1 = 2
print(sys.getrefcount(s2)) # 1 + 1 = 2

s3 = s1
print(sys.getrefcount(s1)) # 2 + 1 = 3
print(sys.getrefcount(s2)) # 2

s4 = s1
print(sys.getrefcount(s1)) # 3 + 1 = 4
print(sys.getrefcount(s2)) # 2

s3 = s2
s4 = s2
print(sys.getrefcount(s1)) # 4 - 2 = 2
print(sys.getrefcount(s2)) # 2 + 2 = 4
```
